```python
import matplotlib.pyplot as plt

# Sample data
labels = ['Real News', 'Fake News']
counts = [3500, 1500]  # Replace with actual values

# Bar positions
x = range(len(labels))

# Create bar chart
plt.bar(x, counts, color=['green', 'red'])
plt.xticks(x, labels)
plt.ylabel('Number of Articles')
plt.title('Real vs Fake News Distribution')

# Add value labels
for i, v in enumerate(counts):
    plt.text(i, v + 50, str(v), ha='center')

plt.tight_layout()
plt.show()
```
